```python
sample.fillna({"sell_price":0},inplace=True)
```

```python
def categorical_analysis(name,sample=sample):
    df_temp = sample.groupby([name,"variable"]).agg({"value":np.sum,"sell_price":np.sum,"id":pd.Series.nunique})
    df_temp["value"] = df_temp["value"]/df_temp["id"]
    df_temp["sell_price"] = df_temp["sell_price"]/df_temp["id"]
    df_temp.reset_index(inplace=True)
    f, ax = plt.subplots(2,1,figsize=(10,8))
    df_temp.sort_values("value",inplace=True)
    sns.boxplot(x=name,y="value",data=df_temp,ax=ax[0])
    sns.boxplot(x=name,y="sell_price",data=df_temp,ax=ax[1])
    plt.subplots_adjust( 
                    wspace=0.4, 
                    hspace=0.4)
    plt.show()
    del df_temp
```

```python
categorical_analysis("dept_id")
```

```python
def is_sold(x):
    x1=np.where(x>1,1,0)
    return sum(x1)
```

```python
def val_check(x,low,high):
    low=int(low.replace("d_",""))
    high=int(high.replace("d_",""))
    x=int(x.replace("d_",""))
    if(x>low and x<high):
        return True
    else:
        return False
```
